```python
import numpy as np
import tensorflow as tf
import tensorflow.keras.datasets.mnist as fation
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout , LSTM , MaxPooling2D
from tensorflow.keras.utils import to_categorical

# load dataset
( X_train , Y_train ) , ( X_test , Y_test ) = fation.load_data()

# normalize dataset
X_train = X_train.astype("float32") / 255
X_test = X_test.astype("float32") / 255

# print(X_test)
# Y_train = to_categorical(Y_train)
# Y_test = to_categorical(Y_test)

# build a model 
model = Sequential()

model.add( LSTM(256 , input_shape=(28,28) , return_sequences=True) )  # return_sequences : return the last output or not
model.add( Dropout(0.2) ) # normalize data by first arg

model.add( LSTM(32) )
model.add( Dropout(0.2) )

model.add( Dense(128 , activation="relu") )
model.add( Dense(10 , activation="softmax") )

# optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

model.compile( loss = "sparse_categorical_crossentropy" , metrics=["accuracy"] )

model.fit( X_train , Y_train, validation_data=(X_test,Y_test) , epochs=4 )


#####
'''
epochs is important beacause make train better
'''

#####
```

```python
test_results = model.predict(X_test)

my_predict = 0
for x in range(10000):
    if np.argmax(test_results[x]) == Y_test[x]:
        my_predict += 1

print("-------------------------------------------")
print( "my model : " + str(my_predict/100)  + "%" )
print("-------------------------------------------")
```
